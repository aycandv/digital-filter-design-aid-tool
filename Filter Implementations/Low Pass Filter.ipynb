# Low Pass Filter Implementation

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

## Function Declerations

```python
def createInput(frequency, timeInterval):
    return signal.square(2 * np.pi * frequency * timeInterval)


def lowPassFilter(inputSignal, alpha):
    output = np.zeros(1000)
    for i in range(0, output.size):
        output[i] = alpha * inputSignal[i] + (1 - alpha) * output[i - 1]
    return output
```

## Execution

* You can change the value of **alpha** to see characteristics of the filter.
  * When **alpha** gets closer to 0, cutoff frequency decreases.
* First argument of the function **createInput** is frequency of input signal in Hz.

```python
alpha = 0.005
frequency = 1000 # Hz
t = np.linspace(0, 0.01, 1000, endpoint=True) #time interval

inputSignal = createInput(frequency, t)
filteredSignal = lowPassFilter(inputSignal, alpha)
```

## Plotting Input & Output Signals

```python
plt.plot(t, inputSignal)
plt.plot(t, filteredSignal)
plt.title('Low Pass Filter Example with Alpha: {}'.format(alpha))
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.legend(["Input Signal", "Filtered Signal"])
plt.ylim(-2, 2)
plt.show()
```
